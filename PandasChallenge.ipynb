# PyCity Schools Analysis

Based upon the information and review of the data, I was able to formulate the following observations:

* The Charter schools had a better overall passing rate than compared to the District schools and the gap was significant within the subject of math. 

* Higher per student spending did not correlate to better scores as the lower spending ranges had better overall reading and math scores. 

* Schools with a lower student count seemed to have better test scores thus we can assume this is due to a lower student to teacher ratio; however, this hypothesis cannot be deemed as correct as we do not have information regarding the classroom sizes. 

* Based upon the data analysis, I would be personally inclined to send my child to a school within the small or medium size than the large size. 

* Charter schools seemed to have better success with their students than compared to the district schools.

```python
import pandas as pd
```

```python
# Loading of data for schools and students
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
schooldata_df = pd.read_csv(school_data)
studentdata_df = pd.read_csv(student_data)

# Combining data files into a single dataset.  
schooldata_complete = pd.merge(studentdata_df, schooldata_df, how="left", on=["school_name", "school_name"])
schooldata_complete.head()
```

# District Summary

```python
#Obtaining total number of unique schools within data
totschools = len(schooldata_complete["school_name"].unique())
totschools
```

```python
# Calculating the total number of students
totstudents = schooldata_complete["Student ID"].count()
totstudents
```

```python
# Calculating the total budget for schools
totbudget = sum(schooldata_complete["budget"].unique())
totbudget
```

```python
# Calculating the avg math score for students
avgmathscore = schooldata_complete["math_score"].mean()
avgmathscore
```

```python
# Calculating the avg reading score for students
avgreadscore = schooldata_complete["reading_score"].mean()
avgreadscore
```

```python
# Calculating the % of students who passed math

passmath = schooldata_complete[(schooldata_complete["math_score"] >= 70)].count()["student_name"]
percpassmath = passmath / float(totstudents) * 100
percpassmath
```

```python
# Calculating the % of students who passed reading

passread = schooldata_complete[(schooldata_complete["reading_score"] >= 70)].count()["student_name"]
percpassread = passread / float(totstudents) * 100
percpassread
```

```python
# Calculating the total % of students who passed both reading and math

totalpassrate = schooldata_complete[
    (schooldata_complete["math_score"] >= 70) & (schooldata_complete["reading_score"] >= 70)
].count()["student_name"]
totalpassrate = totalpassrate /  float(totstudents) * 100
totalpassrate
```

```python
# Creation of dataframe to hold the Districty Summary
distsummary_df = pd.DataFrame({
    "Total Schools": [totschools],
    "Total Students": [totstudents],
    "Total Budget": [totbudget],
    "Average Math Score": [avgmathscore],
    "Average Reading Score": [avgreadscore],
    "Passing Math Percentage": [percpassmath],
    "Passing Reading Percentage": [percpassread],
    "Overall Passing Rate": [totalpassrate]
})

distsummary_df['Total Budget'] = distsummary_df['Total Budget'].map('${:,.2f}'.format)
distsummary_df["Total Students"] = distsummary_df["Total Students"].map("{:,}".format)

distsummary_df.head()
```

# School Summary

```python
# Obtaining the school names and school types
schoolnames = schooldata_df.set_index(["school_name"])["type"]
schoolnames
```

```python
# Calculation of the total number of students per school
schoolgroup = schooldata_complete.groupby(["school_name"])
studbyschool = schoolgroup.size()
studbyschool
```

```python
# Calculation of total budget per school
schoolbudgets = schooldata_complete.groupby(["school_name"]).mean()["budget"]

schoolbudgets
```

```python
# Calculation of school budget per student
perstudbudget = schoolbudgets / studbyschool
perstudbudget
```

```python
# Calculation of average math score per school
avgmathscore = schooldata_complete.groupby(["school_name"]).mean()["math_score"]
avgmathscore
```

```python
# Calculation of average reading score per school
avgreadscore = schooldata_complete.groupby(["school_name"]).mean()["reading_score"]
avgreadscore
```

```python
# Calculation of schools with a passing math avg score
studpassmath = schooldata_complete[(schooldata_complete["math_score"]>=70)]
studpassmathperschool = studpassmath.groupby(["school_name"]).count()["student_name"] / studbyschool * 100
studpassmathperschool
```

```python
# Calculation of schools with a passing reading avg score
studreadmath = schooldata_complete[(schooldata_complete["reading_score"]>=70)]
studpassreadperschool = studreadmath.groupby(["school_name"]).count()["student_name"] / studbyschool * 100
studpassreadperschool
```

```python
# Calculation of overall passing for students who passed both math AND reading
passmathandread = schooldata_complete[(schooldata_complete["reading_score"]>=70) & (schooldata_complete["math_score"]>=70)]
passmathandreadperc = passmathandread.groupby(["school_name"]).count()["student_name"] / studbyschool * 100
passmathandreadperc                                                                                                                                                                                           
```

```python
# Creation of data frame to list the summary of each school
schoolsummary_df = pd.DataFrame({"School Type": schoolnames,
      "Total Students": studbyschool,
      "Total School Budget": schoolbudgets,
      "Per Student Budget": perstudbudget,
      "Average Math Score": avgmathscore,
      "Average Reading Score": avgreadscore,
      "% Passing Math": studpassmathperschool,
      "% Passing Reading": studpassreadperschool,
      "% Overall Passing Rate": passmathandreadperc
                                })

schoolsummary_df["Total School Budget"] = schoolsummary_df["Total School Budget"].map("${:,.2f}".format)
schoolsummary_df["Per Student Budget"] = schoolsummary_df["Per Student Budget"].map("${:,.2f}".format)


schoolsummary_df
```

# Highest-Performing Schools (by % Overall Passing)

```python
# List of highest performing schools based upon the overall passing rate
top_schools = schoolsummary_df.sort_values(["% Overall Passing Rate"],ascending = False).head()
top_schools
```

# Lowest-Performing Schools (by % Overall Passing)

```python
# List of lowest performing schools based upon the overall passing rate
bottom_schools = schoolsummary_df.sort_values(["% Overall Passing Rate"],ascending = True).head()
bottom_schools
```

# Math Scores by Grade 

```python
# Sorting math grades per school and grades 
ninthgrademathscores = schooldata_complete[schooldata_complete["grade"] == "9th"].groupby(["school_name"])["math_score"].mean()
tenthgrademathscores = schooldata_complete[schooldata_complete["grade"] == "10th"].groupby(["school_name"])["math_score"].mean()
eleventhgrademathscores = schooldata_complete[schooldata_complete["grade"] == "11th"].groupby(["school_name"])["math_score"].mean()
twelthgrademathscores = schooldata_complete[schooldata_complete["grade"] == "12th"].groupby(["school_name"])["math_score"].mean()

mathscoresbygrade = pd.DataFrame ({"9th": ninthgrademathscores,
      "10th": tenthgrademathscores,
      "11th": eleventhgrademathscores,
      "12th": twelthgrademathscores})

mathscoresbygrade.index.name = None

mathscoresbygrade
```

# Reading Score by Grade

```python
# Sorting reading grades per school and grades 
ninthgradereadscores = schooldata_complete[schooldata_complete["grade"] == "9th"].groupby(["school_name"])["reading_score"].mean()
tenthgradereadscores = schooldata_complete[schooldata_complete["grade"] == "10th"].groupby(["school_name"])["reading_score"].mean()
eleventhgradereadscores = schooldata_complete[schooldata_complete["grade"] == "11th"].groupby(["school_name"])["reading_score"].mean()
twelthgradereadscores = schooldata_complete[schooldata_complete["grade"] == "12th"].groupby(["school_name"])["reading_score"].mean()

readscoresbygrade = pd.DataFrame ({"9th": ninthgradereadscores,
      "10th": tenthgradereadscores,
      "11th": eleventhgradereadscores,
      "12th": twelthgradereadscores})

readscoresbygrade.index.name = None

readscoresbygrade
```

```python
#Unformat of prior DataFrame with $ to allow pull for per student budget data. I created a new dataframe as I was receiving an error for the final portion of the assignment.
schoolsummaryunformat_df = pd.DataFrame({"School Type": schoolnames,
      "Total Students": studbyschool,
      "Total School Budget": schoolbudgets,
      "Per Student Budget": perstudbudget,
      "Average Math Score": avgmathscore,
      "Average Reading Score": avgreadscore,
      "% Passing Math": studpassmathperschool,
      "% Passing Reading": studpassreadperschool,
      "% Overall Passing Rate": passmathandreadperc
                                })


schoolsummaryunformat_df
```

# Scores by School Spending

```python
# Establishing the bins for the monetary values of spending ranges and scores
spending_bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]

schoolsummaryunformat_df["Spending Ranges (Per Student)"] = pd.cut(schoolsummaryunformat_df["Per Student Budget"], spending_bins, labels=group_names)

spending_math_scores = schoolsummaryunformat_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = schoolsummaryunformat_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = schoolsummaryunformat_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = schoolsummaryunformat_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = schoolsummaryunformat_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing Rate"]

spending_summary = pd.DataFrame({"Average Math Score" : spending_math_scores,
                        "Average Reading Score" : spending_reading_scores,
                        "% Passed Math" : spending_passing_math,
                        "% Passed Reading" : spending_passing_reading,
                        "% Overall Passed" : overall_passing_spending})

spending_summary
```

# Scores by School Size

```python
# Establishing the bins for the size of each school based upon enrollment and scores
size_bins = [0, 1000, 2000, 5000]
group_labels = ["Small (<1000", "Medium (1000-2000)", "Large (2000-5000)"]

schoolsummaryunformat_df["School Size"] = pd.cut(schoolsummaryunformat_df["Total Students"], size_bins, labels=group_labels)

size_math_scores = schoolsummaryunformat_df.groupby(["School Size"]).mean()["Average Math Score"]
size_reading_scores = schoolsummaryunformat_df.groupby(["School Size"]).mean()["Average Reading Score"]
size_passing_math = schoolsummaryunformat_df.groupby(["School Size"]).mean()["% Passing Math"]
size_passing_reading = schoolsummaryunformat_df.groupby(["School Size"]).mean()["% Passing Reading"]
size_overall_passing = schoolsummaryunformat_df.groupby(["School Size"]).mean()["% Overall Passing Rate"]

size_summary = pd.DataFrame({"Average Math Score" : size_math_scores,
                        "Average Reading Score" : size_reading_scores,
                        "% Passed Math" : size_passing_math,
                        "% Passed Reading" : size_passing_reading,
                        "% Overall Passed" : size_overall_passing})

size_summary
```
